결국 이항함수는 팩토리얼의 조합으로 구현해낼 수 있음! 

```python
# factorial 을 함수로 정의해보자

def factorial(n):
    res = 1
    for i in range(1, n+1):
        res *= i
    return res
```

```python
factorial(5)
```

```python
Image("./1st_image/image14.jpg")
```

```python
# 먼저, 팩토리얼 자체를 쓰면 얼마나 긴지
factorial(4)/(factorial(2)*factorial(2))
```

팩토리얼로 코드를 짜도 좋지만, 이항함수 자체를 함수로 만들어 사용해보자

```python
def binomial(n, x):
    res = factorial(n)/(factorial(x)*factorial(n-x))
    return res
```

```python
binomial(4,2)
```
