# 범주형 vs 수치형 변수

```python
plt.figure(figsize = (12,10))
sns.barplot(data=df, x='origin', y='mpg', ci = None)
```

## Groupby를 통한 연산

```python
# 위의 그래프에서, mpg의 값은 default가 mean(평균) 임.
# groupby를 통해 origin 별로 그룹화하고, mpg의 평균을 직접 구해보자

df.groupby('origin').mean()['mpg']
```

# pivot table을 통한 연산

```python
pd.pivot_table(data=df, index='origin', values='mpg')
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=df, x='origin', y='mpg', estimator=np.sum, ci=None)
```

```python
df.groupby(by=['origin', 'cylinders'])['mpg'].sum()
```

```python
# unstack -> 마지막 인덱스의 값을 컬럼으로 만들어주는 것!
df.groupby(by=['origin', 'cylinders'])['mpg'].sum().unstack()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=df, x='origin', y='mpg', hue='cylinders', estimator=np.sum, ci=None)
```

```python
pd.pivot_table(data=df, index='origin', 
               columns='cylinders',
               values='mpg', aggfunc=np.sum)
```

# groupby 와 pivot_table 은 사실 같은 기능!
## pivot_table을 쉽게 사용할 수 있게 만든게 groupby 기능임!
## groupby 를 추상화한 것이 pivot_table 이다!

```python
plt.figure(figsize=(12,10))
sns.boxplot(data=df, x='origin', y='mpg')
```

```python
# groupby로 origin 값에 따른 mpg의 기술통계 구하기
# 결과를 변수에 할당하여 재사용하기
mpd_desc = df.groupby('origin')['mpg'].describe()
mpd_desc
```

```python
europe = mpd_desc.loc["europe"]
europe
```

# boxplot 파헤치기

```python
Q3 = europe["75%"]
Q1 = europe["25%"]
IQR = Q3 - Q1 # 사분위수 범위
OUT_MAX = Q3 + (1.5 * IQR)
OUT_MIN = Q1 - (1.5 * IQR)
Q3, IQR, OUT_MAX, OUT_MIN
```

-  OUT_MAX를 초과하면 별표로 표시,
-  OUT_MIN를 초과하면 별표로 표시!

```python
df_europe = df[df['origin'] == 'europe']
sns.boxplot(data=df_europe, x='mpg')
```

```python
sns.boxenplot(data=df_europe, x='mpg')
```

```python
sns.violinplot(data=df_europe, x='mpg')
```
