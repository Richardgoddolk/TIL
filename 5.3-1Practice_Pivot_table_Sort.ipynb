```python
import pandas as pd
```

```python
fpath = './data/babyNamesUS.csv'
raw = pd.read_csv(fpath)
raw.head()
```

```python
raw.info()
```

```python
# StateCode, Sex, YearOfBirth에 상관없이 이름이 등록된 수를 합하여 정리
# 인덱스 이름, 값은 등록된 수의 총 합으로 피벗테이블 만들기

raw.pivot_table(index = 'Name', values = 'Number', aggfunc = 'sum') 
# aggfunc 의 디폴트 값은 'mean', 평균 이므로 sum 으로 재지정해준다.
```

```python
# 앞의 피벗테이블에서 성별 구분 컬럼에 추가
name_df = raw.pivot_table(index = 'Name', values = 'Number', aggfunc = 'sum')
```
