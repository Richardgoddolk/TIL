```python
columns_select = ['팀', '이름', '생일', '일자', '상대', '타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비']
data = raw[columns_select]
data.head()
```

# KBO 최고의 타자 구하기

선수별 기록 집계

```python
data_player = data.pivot_table(index = ['팀', '이름', '생일'],
                              values = ['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
                              aggfunc = 'sum')
data_player.head()
```

```python
data_player['타수'].hist()

# 많은 선수들이 0~50 사이에 분포한다는 것 알 수 있음!
```

## 타수가 50보다 큰 선수들만 선택,

reset_index() 를 이용해 현재 인덱스로 설정된 팀/이름/생일 데이터를 컬럼으로 옮기기

```python
cond = data_player['타수'] > 50
data_player = data_player[cond].reset_index()
data_player.sample(10)
```

```python
# 타율, 출루율, 장타율, OPS 를 계산하는 함수 만들기

def cal_hit(df):
    
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['볼넷'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    
    return df
```

```python
player_stat = cal_hit(data_player)
player_stat.sample(10)
```

```python
player_stat = player_stat.sort_values( by = ['출루율', '장타율', 'OPS', '타율'], ascending = False)
player_stat = player_stat.reset_index(drop = True)
player_stat.sample(10)
```

```python
player_stat = player_stat.drop(columns = ['index'])
player_stat.sample(10)
```
