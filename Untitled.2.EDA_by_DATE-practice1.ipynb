# 승객이 가장 많은 날짜 찾기

```python
# 날짜별, 요일별 승차총승객수 합계 데이터 정리하기
data_date = raw.pivot_table(index = ['사용일자', '요일'], values = '승차총승객수', aggfunc = 'sum')
data_date.sample(10)
```

## 5월의 승차총승객수가 많은건지, 금요일 승차총승객수가 많은건지 확인해봐야 함!


### 가정1 ) 5월이 승차총승객수가 많은 것이다.
### 가정2 ) 금요일의 승차총승객수가 많은 것이다. 

현재 사용일자만 있고, 월별 컬럼이 없으므로 월별 컬럼을 새로 만들어주자.

여기에서, 연월/월일 컬럼 2가지를 만들어보자

```python
# 현재 사용일자의 데이터는 YYYYMMDD 으로 되어 있음, 이것을 슬라이싱 해서 만들어보자!
yearmonth_list = [ ]
monthday_list = [ ]

for date in raw['사용일자']:
    yearmonth = str(date)[ : 6]
    monthday = str(date)[4 : ]
    yearmonth_list.append(yearmonth)
    monthday_list.append(monthday)
    
raw['연월'] = yearmonth_list
raw['월일'] = monthday_list
```

```python
raw.head()
```

```python
raw['연월'].unique()
```

## 이제, "가정1 ) 5월이 승차총승객수가 많은 것이다."
를 검증해보자!


피벗테이블 만들기

```python
date_month = raw.pivot_table(index = '연월', values = '승차총승객수', aggfunc = 'sum').sort_values(by = '승차총승객수', ascending = False)
date_month
```

### 가정1 결론 : (19년) 5월에 승객수가 많은 것은 사실이다!

## 이제, "가정2 ) 금요일에 승객수가 많다."
를 검증해보자!

피벗테이블 만들기

```python
date_day = raw.pivot_table(index = '요일', values = '승차총승객수', aggfunc = 'sum').sort_values(by = '승차총승객수', ascending = False)
date_day.head()
```

### 가정2 결론 : 금요일의 승객수가 많은 것은 사실이다!

---
## 이제, 월별/일자별 승객 수를 그래프로 살펴보자!

```python
# 먼저, 19년 1월 데이터만 선택해보자!
yearmonth = '201901'
cond = raw['연월'] == yearmonth
df_selected = raw[cond]
df_selected.head(10).sort_values(by = '승차총승객수', ascending = False)
```

```python
# 위의 19년 1월 데이터를 일자/요일별 데이터로 정리!

date_day = df_selected.pivot_table(index = ['월일', '요일'], values = '승차총승객수', aggfunc = 'sum')
date_day
```

```python
# 그래프 그리기 전, 데이터 프레임에서 월일과 요일이 index 값으로 들어가 있으면 그래프의 축으로 설정이 안된다! 
# 따라서 인덱스를 모두 컬럼으로 넘겨주자
date_day = date_day.reset_index()
date_day
```
