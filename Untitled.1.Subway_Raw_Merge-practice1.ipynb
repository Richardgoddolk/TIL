### Q) 요일을 추가?

```python
from datetime import datetime
```

```python
# strptime : 문자를 날짜형식으로 바꾸는 명령어!

s = '2022-06-14'
datetime.strptime(s, '%Y-%m-%d')
```

```python
s = '20220613'
date = datetime.strptime(s, '%Y%m%d')
date
```

```python
weekday = date.weekday()
weekday
# 월요일부터 0, 일요일은 6!
```

```python
weekday_dict = ['월', '화', '수', '목', '금', '토', '일']
weekday_index = date.weekday()

print(weekday_index) # 위에서 s 로 지정했던 날짜의 인덱스(월~일 : 0~6 숫자 부여)
print(weekday_dict[weekday_index]) # weekday 리스트의 번호에 맞추어 요일을 변환한다
```

```python
# 요일 컬럼을 추가하기 위한 본격적인 작업!
# weekday 리스트를 먼저 만들어 놓고, 사용일자 컬럼을 문자로 가져온 뒤, 날짜 형식으로 전환, datetime 라이브러리의 weekday 함수를 이용해 요일로 만들기!

weekday_list = [ ]

for date_str in raw['사용일자'] :
    
    date = datetime.strptime(str(date_str), '%Y%m%d')
    
    weekday_index = date.weekday()
    weekday = weekday_dict[weekday_index]
    
    weekday_list.append(weekday)
    
weekday_list
```

```python
raw['요일'] = weekday_list
raw.sample(10)
```
