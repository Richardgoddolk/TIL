```python
import pandas as pd

file = './data/KBO_2019_player_gamestats.csv'
raw = pd.read_csv(file, encoding = 'cp949')
```

```python
raw.head()
```

```python
raw.columns
```

```python
# 월별 데이터를 뽑기 위해 '월-일' 데이터 형식으로 이루어진 '일자' 컬럼이 필요하다!

month_list = [ ]

for monthdate in raw['일자']: # '일자' 컬럼의 데이터를 monthdate 변수에 하나씩 넣어줌
    
    month = monthdate.split('-')[0] # monthdate 변수의 데이터에 일자가 들어가있는데, 이걸 '-'기호를 중심으로 나눈다
    month_list.append(month)
    
raw['월'] = month_list
raw.head()
```

```python
# 위의 수많은 컬럼 중 필요한 컬럼만 사용!

columns_select =  ['팀', '이름', '생일','일자', '상대','타수','안타','홈런', '루타', '타점','볼넷', '사구', '희비', '월']

data = raw[columns_select]
data.head()
```

```python
# 이제 여기서 월별 실적을 피벗테이블로 집계해보자!

data_player_month = data.pivot_table(index = ['팀', '이름', '생일', '월'],
                                     values = ['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
                                     aggfunc = 'sum'
                                    )
data_player_month.head()
```

```python
# 위에서 한 선수의 월별 실적 데이터가 이름 하나에 월 전체가 묶여 있으므로, 월별 실적을 비교하기엔 어렵다!
# 따라서 월별로 index를 풀어주자.

data_player_month = data_player_month.reset_index()
data_player_month

# 월별로 데이터가 한 로우로 나눠졌다!!!
```

```python
def cal_hit(df):
    '''
    - 타율 : 공을 쳐서 나가는 비율 --> 안타 / 타수
    - 출루율: 진루해서 나가는 비율 -->  (안타+볼넷+몸에맞는볼)/(타수+볼넷+몸에맞는볼+희생플라이)
    - 장타율 : 타율에 진루한 베이스 가중치 추가 -->   루타 / 타수
    '''
    
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    return df
```

```python
# data_player_month  데이터의 타자별 주요 실적을 계산합니다. 
player_month_stat = cal_hit(data_player_month)
player_month_stat = player_month_stat.dropna() # player_month_sate 부분에서 어느 축에서도 데이터가 없는 해당 행을 삭제
player_month_stat
```
