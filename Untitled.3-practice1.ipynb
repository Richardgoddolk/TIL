```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import font_manager, rc
import platform
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use(['dark_background'])
sns.set(rc = {'figure.figsize':(15,8)})

# 이미지 한글 표시 설정
if platform.system() == 'Windows':  # 윈도우인 경우 맑은고딕
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:    # Mac 인 경우 애플고딕
    rc('font', family='AppleGothic')

#그래프에서 마이너스 기호가 표시되도록 하는 설정입니다.
matplotlib.rcParams['axes.unicode_minus'] = False   
```

```python
file = './data/KBO_2019_player_gamestats.csv'
raw = pd.read_csv(file, encoding = 'cp949')
raw.head(10)
```

```python
raw['상대'].unique()
# @가 붙은 경우에는 원정경기, 없으면 홈 경기 의미! 
# @ 표시를 없애고, 새로운 컬럼에 홈/원정 경기에 대한 정보를 만들어보자!
# '상대' 컬럼을 가지고 홈/원정 여부, 상대팀을 분리하여 각각 '홈어웨이', '상대팀'컬럼으로 따로 저장
```

```python
opp_list = [ ]
home_away_list = [ ]

for opp in raw['상대']:
    
    if '@' in opp :
        home_away = '원정'
        opp = opp.replace('@', '')
    
    else :
        home_away = '홈'
        
    opp_list.append(opp)
    home_away_list.append(home_away)
    
raw['홈어웨이'] = home_away_list
raw['상대팀'] = opp_list

raw.sample(10)
```

```python
# 상대 팀별 실적을 정리, 피벗 테이블 만들기!
# 상대팀별 실적을 정리

data = raw.pivot_table(index =['팀', '이름', '생일', '상대팀'],  # '상대' 컬럼으로 인덱스를 지정할 경우 @가 붙은것까지 포함되므로, 방금전에 따로 지정한 '상대팀' 컬럼을 이용하도록 한다!
                values = ['타수','안타','홈런', '루타', '타점','볼넷', '사구', '희비'],
                aggfunc = 'sum')

data.head()
```

```python
cond = data['타수'] > 0 
data = data[cond]
data.head()
```

```python

data = data.reset_index()  
data.head()
```

```python
def cal_hit(df):
    '''
    - 타율 : 공을 쳐서 나가는 비율 --> 안타 / 타수
    - 출루율: 진루해서 나가는 비율 -->  (안타+볼넷+몸에맞는볼)/(타수+볼넷+몸에맞는볼+희생플라이)
    - 장타율 : 타율에 진루한 베이스 가중치 추가 -->   루타 / 타수
    '''
    
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    return df
```

```python
player_stats_opp = cal_hit(data)
player_stats_opp
```

# 특정팀 킬러가 있는가?

```python
# 두산을 상대로 타수가 좋은 선수 뽑아내기
team = '두산'
cond = (player_stats_opp['상대팀'] == team) & (player_stats_opp['타수'] > 10)
player_stats_opp[cond].sort_values(by = '출루율', ascending = False).head(10)

```

```python
team = '롯데'
cond = (player_stats_opp['상대팀'] == team) & (player_stats_opp['타수'] > 20)
player_stats_opp[cond].sort_values(by = '출루율', ascending = False).head(10) # 이렇게 하면 팀별로 상위 10명의 데이터가 들어감! 
```

```python
# 전체 팀을 넣고 상대팀을 상대로 높은 성과를 보여준 선수 추출해내기
hitter_df = pd.DataFrame()

for team in player_stats_opp['상대팀'].unique() : 
    
    cond = (player_stats_opp['상대팀'] == team) & (player_stats_opp['타수'] > 20)
    df = player_stats_opp[cond].sort_values(by = '출루율', ascending = False).head()
    
    hitter_df = hitter_df.append(df)

hitter_df
```

### 특정팀 상대 출루율 상위선수 데이터프레임 만드는 방법1
