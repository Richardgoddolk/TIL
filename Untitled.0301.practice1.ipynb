## 결측치 비율 구하기

### data가 있는 경우엔 True, 없는 경우엔 False로 구분된다! 
### 여기서 (true + true + true + false) / 4 를 그냥 계산하면 0.75 -> 즉, 데이터가 있는 경우의 비율이 나오는데, 
### 결측치의 비율을 구할 때는 isnull().mean() 을 하면 위에서 True 대신 False 값으로 비율이 나오게 되는 것!

```python
(True + True + False + True) / 4
```
